```python
def my_function(y, w, x, b):
    final_result = 0
    final_result = sigmoid(np.dot(w,x.T)+b)
    cost = (-1/14000)*(np.sum((y.T*np.log(final_result)) + ((1-y.T)*(np.log(1-final_result)))))
    dw = 0
    dw = (1/14000)*(np.dot(x.T, (final_result-y.T).T))
    db = 0
    db = (1/14000)*(np.sum(final_result-y.T))
    grads = {"dw": dw, "db": db}
    return grads, cost
```
