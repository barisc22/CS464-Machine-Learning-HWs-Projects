```python
def mutinomial(array_list, count):
    j = 0
    p = 11443
    n = count
    hop = []
    while(j < p):
        i = 0
        sum = 0
        while(i < n):
            sum = sum + int(array_list[i][j])
            i = i + 1
        hop.append(sum)
        j = j + 2
    
    print(hop)
    return hop
```

```python
def bernoulli(array_list, count):
    j = 0
    p = 11443
    n = count
    hop = []
    while(j < p):
        i = 0
        sum = 0
        while(i < n):
            if(int(array_list[i][j]) > 0):
                sum = sum + 1
            i = i + 1
        hop.append(sum)
        j = j + 2
    
    print(hop)
    return hop
```
